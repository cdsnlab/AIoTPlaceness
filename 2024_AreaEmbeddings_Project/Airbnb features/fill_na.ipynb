```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('../Dataset/AirBnB_LLM/gemma2_no_listing_raw.csv')
df
```

```python
dim = len(ast.literal_eval(df['LLM Embeddings'][0]))   # 임베딩 크기 확인
dim
```

```python
# 날짜 범위 생성
date_range = pd.date_range(start='2017-01-01', end='2022-07-01', freq='MS')
date_range = date_range.strftime('%Y-%m-%d').tolist()  # 문자열로 변경
print(len(date_range))

# 동 이름과 날짜 범위의 모든 조합 생성
dong_names = list(df['ADM_NM'].unique())
print(len(dong_names))

full_index = pd.MultiIndex.from_product([dong_names, date_range], names=['ADM_NM', 'Reporting Month'])
#full_index

# 기존 데이터프레임에서 인덱스 생성
existing_index = pd.MultiIndex.from_frame(df[['ADM_NM', 'Reporting Month']])
#existing_index

# 존재하지 않는 조합 찾기
missing_combinations = set(full_index) - set(existing_index)

len(missing_combinations)
```

```python
# 누락된 부분 값 채우기
missing_combinations_df = pd.DataFrame(
    list(missing_combinations), 
    columns=['ADM_NM', 'Reporting Month']
)

zero_vector = [0]*dim
print(zero_vector)

missing_combinations_df['Property ID'] = 'x'
fills = ['Occupancy Rate', 'Revenue (USD)',	'Number of Reservations', 'LLM Embeddings']
for c in fills:
    if c == 'LLM Embeddings':
        missing_combinations_df[c] = [zero_vector] * len(missing_combinations_df)
    else:
        missing_combinations_df[c] = 0
    
missing_combinations_df['Reporting Month'] = missing_combinations_df['Reporting Month'].astype(str)

missing_combinations_df
```

```python
result = pd.concat([df, missing_combinations_df])
result
```

```python
result = result[result['Reporting Month'].isin(date_range)]
# Convert 'Reporting Month' to datetime format
result['Reporting Month'] = pd.to_datetime(result['Reporting Month'], format='%Y-%m-%d')

# Sort by 'Reporting Month' and then 'ADM_NM'
result = result.sort_values(by=['Reporting Month', 'ADM_NM'], ascending=[True, True])

# Now df_sorted is sorted first by month and then by ADM_NM within each month
```

```python
result.to_csv('../Dataset/AirBnB_LLM/gemma2_no_listing_raw_fill_na_ver.csv', index=False, encoding='utf-8-sig')
```
