```python
df_bigloc = pd.read_csv(os.path.join('./csv', 'bigloc-loc.tsv'), header=None, sep='\t', encoding='utf-8-sig')
```

```python
print(df_bigloc[:50])
```

```python
station_set = set()
location_dict = {}
for index, row in df_bigloc.iterrows():
    station_name = row.iloc[0]
    location_id = row.iloc[1]
    if station_name not in station_set:
        station_set.add(station_name)
    location_dict[location_id] = station_name
print(location_dict)
print(station_set)
```

```python
df_data = pd.read_csv(os.path.join(CONFIG.TARGET_PATH, 'SEOUL_SUBWAY_DATA-3.csv'), index_col=1,
                          encoding='utf-8-sig')
```

```python
df_data[:5]
```

```python
shortcode_station = {}
for index, row in df_data.iterrows():
    location_id = row.iloc[3]
    station_name = location_dict[location_id]
    shortcode_station[index] = station_name
```

```python
df_mdec_label = pd.read_csv(os.path.join('./csv', 'mdec_label.csv'), index_col=0, encoding='utf-8-sig')
```

```python
print(df_mdec_label[:5])
label_array = np.array(df_mdec_label['labels'])
print(np.max(label_array))
```

```python
station_label_dict = {}
for station_name in station_set:
    station_label_dict[station_name] = np.zeros(12)
print(station_label_dict)
```

```python
for index, row in df_mdec_label.iterrows():
    post_label = row.iloc[0]
    station_name = shortcode_station[index]
    station_label_dict[station_name][post_label] = station_label_dict[station_name][post_label] + 1
print(station_label_dict)
```

```python
categories = ['Eating and drinking', 'Arts and Entertainment', 'Socializing and Communicating', 'Traveling',
                  'Personal care', 'Sports, exercise, and recreation', 'Relaxing and Leisure', 'Consumer purchases',
                  'Attending or Hosting Social Events', 'Working and work-related activities', 'Professional and personal care services', 'Educational activities']
df_station_label = pd.DataFrame.from_dict(station_label_dict, orient='index', columns=categories)
df_station_label.sort_index(inplace=True)
df_station_label = df_station_label.T
print(df_station_label)
```

```python
new_columns = [x[3:] for x in df_station_label.columns]
print(new_columns)
df_station_label.columns = new_columns
print(df_station_label)
```
